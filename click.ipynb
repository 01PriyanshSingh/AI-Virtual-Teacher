```python
import pyautogui
import time
import keyboard

def auto_click(delay=0.1, start_delay=5):
    print(f"Waiting for {start_delay} seconds before starting...")
    time.sleep(start_delay)
    print("Starting auto-clicker. Press 'q' to stop.")
    try:
        while True:
            pyautogui.click()
            time.sleep(delay)
            if keyboard.is_pressed('q'):
                print("Detected 'q'. Exiting auto-clicker.")
                break
    except KeyboardInterrupt:
        print("Auto clicker interrupted manually.")

if __name__ == "__main__":
    auto_click()
```
